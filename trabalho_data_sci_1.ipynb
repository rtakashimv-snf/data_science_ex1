Nome do Dataset: Índice Geral de Reclamações - IGR
Link: https://dados.gov.br/dados/conjuntos-dados/descontinuado-indice-geral-de-reclamacoes---igr

Orgão responsável: Agência Nacional de Saúde Suplementar

Descrição: O IGR têm como base os dados de beneficiários e reclamações recebidos pela ANS e permitem comparar a atuação das empresas que atuam no setor de Saúde Suplementar.

Número de registros e colunas: 9179, 17

Variaveis importantes

Razão Social (Registro ANS): Razão social e Registro ANS da operador

Cobertura:  Cobertura assistencial da operadora

Porte:  Porte da operadora de acordo com a quantidade de beneficiário

Variaveis de meses referentes aos meses citados

```python
import pandas as pd

df = pd.read_csv("IGR_2022.csv", sep=";", encoding="latin1")

df.shape
```

```python
df.head(25)
```

```python
meses = ["Jan/22","Fev/22","Mar/22","Abr/22",
         "Mai/22","Jun/22","Jul/22","Ago/22",
         "Set/22","Out/22"]

for col in meses:
    df[col] = pd.to_numeric(
        df[col].str.replace(",", ".", regex=False),
        errors="coerce"
    )
```

```python
meses = ["Jan/22","Fev/22","Mar/22","Abr/22",
         "Mai/22","Jun/22","Jul/22","Ago/22",
         "Set/22","Out/22"]
```

# **Media**

```python
media = df[meses].mean()
print(f'MEDIA: {media}')
```

# **Mediana**

```python
mediana = df[meses].median()
print(f'MEDIANA: {mediana}')
```

# **Moda**

```python
moda = df[meses].mode().iloc[0]
print(f'MODA: {moda}')
```

# **Minimo**

```python
minimo = df[meses].min()
print(f'MIN: {minimo}')
```

# **Maximo**

```python
maximo = df[meses].max()
print(f'MAX: {maximo}')
```

# **Desvio Padrão**

```python
desvio_padrao = df[meses].std()
print(f'DP: {desvio_padrao}')
```

# **Variância**

```python
variancia = df[meses].var()
print(f'VARIANCIA: {variancia}')
```

# **Quartil 1**

```python
q1 = df[meses].quantile(0.25)
print(f'Q1: {q1}')
```

# **Quartil 2**

```python
q2 = df[meses].quantile(0.50)  # mediana
print(f'Q2: {q2}')
```

# **Quartil 3**

```python
q3 = df[meses].quantile(0.75)
print(f'Q3: {q3}')
```

# **Intervalo interquartil**

```python
iqr = q3 - q1
print(f'IQR: {iqr}')
```

# **Valores Ausentes**

```python
valores_ausentes = df[meses].isnull().sum()
print(f'VALORES AUSENTES: {valores_ausentes}')
```

# **Tipos de dados**

```python
tipos = df.dtypes
```

```python
estatisticas = pd.DataFrame({
    "Média": media,
    "Mediana": mediana,
    "Moda": moda,
    "Mínimo": minimo,
    "Máximo": maximo,
    "Desvio Padrão": desvio_padrao,
    "Variância": variancia,
    "Q1": q1,
    "Q3": q3,
    "IQR": iqr,
    "Valores Ausentes": valores_ausentes
})

print(estatisticas)
```

```python
import matplotlib.pyplot as plt
```

# **Histogramas**

```python
for col in meses:
    plt.figure()
    df[col].hist(bins=30)
    plt.title(f"Histograma - {col}")
    plt.xlabel("IGR")
    plt.ylabel("Frequência")
    plt.show()
```

# **Boxplots**

```python
plt.figure()
df[meses].boxplot()
plt.xticks(rotation=45)
plt.title("Boxplot - IGR por mês")
plt.ylabel("IGR")
plt.show()
```
